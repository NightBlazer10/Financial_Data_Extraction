## Summarising DataFrame and Data Cleaning

```python
bs_df.dtypes
```

```python
bs_df = bs_df.drop(0)
pl_df = pl_df.drop(0)
cs_df = cs_df.drop(0)
```

```python
bs_df  = bs_df.set_index(bs_df.columns[0])
pl_df  = pl_df.set_index(pl_df.columns[0])
cs_df  = cs_df.set_index(cs_df.columns[0])
```

```python
bs_df = bs_df.replace(',' , '' , regex = True)
pl_df = pl_df.replace(',' , '' , regex = True)
cs_df = cs_df.replace(',' , '' , regex = True)
```

```python
bs_df = bs_df.replace('--' , '' , regex = True)
pl_df = pl_df.replace('--' , '' , regex = True)
cs_df = cs_df.replace('--' , '' , regex = True)
```

## Changing Data Type of The DataFrame

```python
bs_df1 = pd.DataFrame()
for cols in bs_df.columns:
    temp = pd.to_numeric(bs_df[cols])
    bs_df1 = pd.concat([bs_df1 , temp] , axis = 1)

bs_df1
```

```python
pl_df1 = pd.DataFrame()
for cols in bs_df.columns:
    temp = pd.to_numeric(pl_df[cols])
    pl_df1 = pd.concat([pl_df1 , temp] , axis = 1)

pl_df1
```

```python
cs_df1 = pd.DataFrame()
for cols in cs_df.columns:
    temp = pd.to_numeric(cs_df[cols])
    cs_df1 = pd.concat([cs_df1 , temp] , axis = 1)

cs_df1
```
